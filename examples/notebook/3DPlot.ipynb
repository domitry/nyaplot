```ruby
require 'nyaplot'
require 'nyaplot3d'
```

```ruby
Nyaplot.init_iruby
```

```ruby
x=[];y=[];z=[]
-10.step(10, 0.5) do |i|
  -10.step(10, 0.5) do |j|
    x.push(i)
    y.push(j)
    z.push(Math.sin(Math.sqrt(i*i+j*j))/Math.sqrt(i*i+j*j))
  end
end
z.map!{|val| next (val.nan? ? 0 : val)} #(0,0) will be -inf
```

# Wireframe

```ruby
plot = Nyaplot::Plot3D.new
plot.add(:wireframe, x, y, z)
plot.show
```

# Surface

```ruby
plot = Nyaplot::Plot3D.new
plot.add(:surface, x, y, z)
plot.show
```

# Line

```ruby
step_num = 10000;
p = 10; r = 28; b = 8/3; dt = 0.01

fx = Proc.new{|x,y,z| ((-1)*p*x + p*y)};
fy = Proc.new{|x,y,z| ((-1)*x*z + r*x - y)};
fz = Proc.new{|x,y,z| (x*y - b*z)};

x_arr=[]; y_arr=[]; z_arr=[]
x = 1; y = 1; z = 1
step_num.times do |i|
  x += dt * fx.call(x,y,z);
  y += dt * fy.call(x,y,z);
  z += dt * fz.call(x,y,z);
  x_arr.push(x);
  y_arr.push(y);
  z_arr.push(z);
end

plot = Nyaplot::Plot3D.new
plot.add(:line, x_arr, y_arr, z_arr)
plot.show
```
